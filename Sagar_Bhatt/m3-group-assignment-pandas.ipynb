### üéØ Task 4: Monthly number of transactions and total amount spent

- For each unique "YEAR_MONTH" value in the DataFrame, calculate the following using `pd.groupby()` and `agg()`:
  - "AMOUNT": The total amount spent in that month.
  - "TRANSACTION_COUNT": The number of transactions made in that month.
- Store the results in a new DataFrame named `df_year_month` with the following columns:
  - "YEAR_MONTH"
  - "AMOUNT"
  - "TRANSACTION_COUNT"
- All three columns should be regular columns (not index).
  - `df_year_month.columns` should print `Index(['YEAR_MONTH', 'AMOUNT', 'TRANSACTION_COUNT'], dtype='object')`.
- Sort the DataFrame by "YEAR_MONTH" in ascending order.


````python
# YOUR CODE BEGINS

df_year_month = (
    df.groupby("YEAR_MONTH")
    .agg(
        AMOUNT=("AMOUNT","sum"),
        TRANSACTION_COUNT=("TRANSACTION_DATE", "count")
    )
    .reset_index()
)
    
df_year_month = df_year_month.sort_values("YEAR_MONTH").reset_index(drop=True)

# YOUR CODE ENDS

display(df_year_month.head(3))
print(df_year_month.columns)
````

### üéØ Task 7: Identify potential split transactions

According to City of Denton's Procurement Policy:

> Unless formally approved through another procurement method, P-Cards may be used only for "Micro Purchases" (goods or services up to $3,000).

We want to identify potential split transactions, which occur when a purchase that exceeds $3,000 is divided into multiple smaller transactions (often on the same day) to avoid procurement rules.

- Using `df.groupby()` and `agg()` to identify groups of transactions that meet **all** of the following criteria:
  - Same **"TRANSACTION_DATE"**
  - Same **"CARDHOLDER_NAME"**
  - Same **"MERCHANT"**
  - **More than 1** transaction in the group (`TRANSACTION_COUNT > 1`)
  - The **"COMBINED_AMOUNT"** in the group exceeds **$3,000**
- Create a DataFrame named `df_split_transactions` that contains these grouped results.
- The resulting DataFrame must contain **exactly five columns**:
  - `"TRANSACTION_DATE"`
  - `"CARDHOLDER_NAME"`
  - `"MERCHANT"`
  - `"COMBINED_AMOUNT"`
  - `"TRANSACTION_COUNT"`
- All five fields must be regular columns, not part of the DataFrame index.
- Sort `df_split_transactions` by **"COMBINED_AMOUNT"** in descending order, and then by **"TRANSACTION_COUNT"** in descending order.

#### üßæ Expected Output

Your `df_split_transactions` DataFrame should look like the following (the index column omitted for clarity):

| TRANSACTION_DATE | CARDHOLDER_NAME      | MERCHANT                | COMBINED_AMOUNT | TRANSACTION_COUNT |
| :--------------- | :------------------- | :---------------------- | --------------: | ----------------: |
| 2024-05-03       | Katherine Barnett    | LOWES                   |         34330.0 |                 2 |
| 2024-07-11       | Katherine Barnett    | LOWES                   |         17185.0 |                 3 |
| 2024-08-23       | Carol Martin-Roberts | ATL CENTENNIAL PARK ON  |         11180.7 |                 9 |
| 2022-08-31       | Chaunesy Baker       | ALOFT BY MARRIOTT       |         10357.3 |                 3 |
| 2023-12-12       | Patti Mankin         | USC VITERBI FCCC ONLINE |          9000.0 |                 5 |

```
Index(['TRANSACTION_DATE', 'CARDHOLDER_NAME', 'MERCHANT', 'COMBINED_AMOUNT',
       'TRANSACTION_COUNT'],
      dtype='object')
```


````python
# YOUR CODE BEGINS

df_split_transactions = df.groupby(
    ['TRANSACTION_DATE', 'CARDHOLDER_NAME', 'MERCHANT']
).agg(
    COMBINED_AMOUNT=('AMOUNT', 'sum'),
    TRANSACTION_COUNT=('TRANSACTION_DATE', 'count')
).reset_index()

df_split_transactions = df_split_transactions.query(
    "(TRANSACTION_COUNT > 1) and (COMBINED_AMOUNT > 3000)"
)

df_split_transactions = df_split_transactions.sort_values(
    ["COMBINED_AMOUNT", "TRANSACTION_COUNT"], ascending=False
).reset_index(drop=True)

# YOUR CODE ENDS

display(df_split_transactions.head())
print(df_split_transactions.columns)
````
